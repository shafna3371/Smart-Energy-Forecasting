```python
import pandas as pd
import requests

# 1. Load your dataset
df = pd.read_csv("updated.csv")

# Make sure date column is datetime
df["date"] = pd.to_datetime(df["date"], format="%d-%m-%Y")

# 2. Define location (example: Kerala)
latitude = 10.0
longitude = 76.5

# 3. Get date range from dataset
start_date = df["date"].min().strftime("%d%m%Y") # Corrected date format to YYYYMMDD
end_date = df["date"].max().strftime("%d%m%Y")   # Corrected date format to YYYYMMDD

# 4. NASA POWER API URL
url = (
    "https://power.larc.nasa.gov/api/temporal/daily/point"
    f"?parameters=T2M,RH2M"
    f"&start={start_date}&end={end_date}"
    f"&latitude={latitude}&longitude={longitude}"
    "&community=RE"
    "&format=JSON"
)

response = requests.get(url).json()

# 5. Convert NASA data to DataFrame
weather_data = response["properties"]["parameter"]

weather_df = pd.DataFrame({
    "date": pd.to_datetime(list(weather_data["T2M"].keys())),
    "Temperature": weather_data["T2M"].values(),
    "Humidity": weather_data["RH2M"].values()
})

# 6. Merge with your dataset
final1_df = pd.merge(df, weather_df, on="date", how="left")

# 7. Save final dataset
final1_df.to_csv("updated.csv", index=False)

print("Weather data added successfully!")
from google.colab import files
files.download('electricity_final.csv')
```
