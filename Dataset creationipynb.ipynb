```python

    import pandas as pd
    import numpy as np
    from datetime import datetime, timedelta

    np.random.seed(42)
    n_days = 2192 #6yrs

    base_equipments = {
         'ac': 1200, 'washing-machine': 160, 'water pump': 1000, 'induction-stove': 2000,'mixi':1000,'iron-box':1000
    }
    equip_power={'light_1': 20 , 'fan_1': 60, 'light_2': 20 , 'fan_2': 60,'light_3': 20 , 'fan_3': 60,'light_4': 20 , 'fan_4': 60,'light_5':60 , 'fridge': 230 ,'tv': 120}
    temp_ranges = {
        1: (17, 26),  2: (18, 28),  3: (29, 33),
        4: (30, 36),  5: (29, 36),
        6: (20, 30), 7: (20, 30), 8: (20, 30),
        9: (20, 30), 10: (24, 31), 11: (24, 31),
        12: (18, 25)
    }


    last_month = (None, None)
    SLABS = [
    (50, 3.35),
    (50, 4.25),
    (50, 5.35),
    (50, 7.2),
    (50, 8.5),
    (float('inf'), 9.50)   # above 250
]

    x=0
    residents = 4
    num_equip = 17
    data = []
    monthly_unit = {}
    monthly_bill = {}
    month_unit = 0
    for day in range(n_days):
        total_unit=0
        mn_month_unit = 0
        ev_month_unit = 0
        nt_month_unit = 0
        mn_equip_power={}
        ev_equip_power={}
        nt_equip_power={}
        mn_hrs={}
        ev_hrs={}
        nt_hrs={}
        date_obj = datetime(2019, 1, 1) + timedelta(days=day)
        date_str = date_obj.strftime('%Y-%m-%d')
        day_name = date_obj.strftime('%A')
        if(day_name=='Sunday' or day_name=='Saturday'):
          weekend=1
        else:
          weekend=0


        low, high = temp_ranges[date_obj.month]
        temp = np.random.uniform(low, high)

        if(temp>24 and temp<=30):
            if(weekend):
                  for eqip,power in equip_power.items():
                        if eqip=='fridge':
                          mn_equip_power[eqip] = power * round(np.random.uniform(5,6),2)
                        elif eqip=='light_5':
                          mn_equip_power[eqip] = 0
                        else:
                          mn_equip_power[eqip] = power * round(np.random.uniform(5,6),2)
                        mn_hrs[eqip]=mn_equip_power[eqip]/power
                  for eqip,power in base_equipments.items():
                    if(eqip=='ac'):
                      mn_equip_power[eqip] = 0
                    else:
                      mn_equip_power[eqip] = power * round(np.random.uniform(0,1),2)
                    mn_hrs[eqip]= mn_equip_power[eqip]/power

                  for eqip,power in equip_power.items():
                    if eqip=='fridge':
                      ev_equip_power[eqip] = power * round(np.random.uniform(2,3),2)
                    else:
                      ev_equip_power[eqip] = power * round(np.random.uniform(0,4),2)
                    ev_hrs[eqip]=ev_equip_power[eqip]/power
                  for eqip,power in base_equipments.items():
                    if(eqip=='ac'):
                      ev_equip_power[eqip] = power * round(np.random.uniform(0,1),2)
                    else:
                      ev_equip_power[eqip] = power * round(np.random.uniform(0,0.5),2)
                    ev_hrs[eqip]= ev_equip_power[eqip]/power

                  for eqip,power in equip_power.items():
                    if(eqip=='light_1'  or eqip=='light_2'  or eqip=='light_3'  or eqip=='light_4' or eqip=='fan_4'  or eqip=='tv' ):
                      nt_equip_power[eqip] = 0
                    elif eqip=='fridge':
                      nt_equip_power[eqip] = power * round(np.random.uniform(4,5),2)
                    elif eqip=='light_5':
                      nt_equip_power[eqip] = power * round(np.random.uniform(7,8),2)
                    else:
                      nt_equip_power[eqip] = power * round(np.random.uniform(0,8),2)
                    nt_hrs[eqip]=nt_equip_power[eqip]/power
                  for eqip, power in base_equipments.items():
                    nt_equip_power[eqip] = 0
                    nt_hrs[eqip] = 0
            else:

              for eqip,power in equip_power.items():
                if eqip=='fridge':
                  mn_equip_power[eqip] = power * round(np.random.uniform(4,5),2)
                elif eqip=='light_5':
                  mn_equip_power[eqip] = 0
                else:
                  mn_equip_power[eqip] = power * round(np.random.uniform(0,5),2)
                mn_hrs[eqip]=mn_equip_power[eqip]/power
              for eqip,power in base_equipments.items():
                if(eqip=='ac'):
                  mn_equip_power[eqip] = 0
                else:
                  mn_equip_power[eqip] = power * round(np.random.uniform(0,1),2)
                mn_hrs[eqip]= mn_equip_power[eqip]/power

              for eqip,power in equip_power.items():
                if eqip=='fridge':
                  ev_equip_power[eqip] = power * round(np.random.uniform(2,3),2)
                else:
                  ev_equip_power[eqip] = power * round(np.random.uniform(0,4),2)
                ev_hrs[eqip]=ev_equip_power[eqip]/power
              for eqip,power in base_equipments.items():
                if(eqip=='ac'):
                  ev_equip_power[eqip] = power * round(np.random.uniform(0,1),2)
                else:
                  ev_equip_power[eqip] = power * round(np.random.uniform(0,0.5),2)
                ev_hrs[eqip]= ev_equip_power[eqip]/power

              for eqip,power in equip_power.items():
                if(eqip=='light_1'  or eqip=='light_2'  or eqip=='light_3'  or eqip=='light_4' or eqip=='fan_4'  or eqip=='tv' ):
                  nt_equip_power[eqip] = 0
                elif eqip=='fridge':
                  nt_equip_power[eqip] = power * round(np.random.uniform(4,5),2)
                elif eqip=='light_5':
                  nt_equip_power[eqip] = power * round(np.random.uniform(7,8),2)
                else:
                  nt_equip_power[eqip] = power * round(np.random.uniform(0,8),2)
                nt_hrs[eqip]=nt_equip_power[eqip]/power
              for eqip, power in base_equipments.items():
                nt_equip_power[eqip] = 0
                nt_hrs[eqip] = 0


        elif(temp>30):

                for eqip ,power in equip_power.items():
                  if eqip=='fan_3' or eqip=='fan_4':
                    mn_equip_power[eqip] = power * round(np.random.uniform(4,7),2)
                  elif eqip=='fridge':
                    mn_equip_power[eqip] = power * round(np.random.uniform(5,6),2)
                  elif eqip=='light_5':
                    mn_equip_power[eqip] = 0
                  else:
                    mn_equip_power[eqip] = power * round(np.random.uniform(0,5),2)
                  mn_hrs[eqip]=mn_equip_power[eqip]/power
                for eqip,power in base_equipments.items():
                  if(eqip=='ac'):
                    mn_equip_power[eqip] = power * round(np.random.uniform(0,2),2)
                  else:
                    mn_equip_power[eqip] = power * round(np.random.uniform(0,1),2)
                  mn_hrs[eqip]=mn_equip_power[eqip]/power

                for eqip,power in equip_power.items():
                  if eqip=='fridge':
                    ev_equip_power[eqip] = power * round(np.random.uniform(2,3),2)
                  else:
                    ev_equip_power[eqip] = power * round(np.random.uniform(0,4),2)
                  ev_hrs[eqip]= ev_equip_power[eqip]/power
                for eqip,power in base_equipments.items():
                  if(eqip=='ac'):
                    ev_equip_power[eqip] = 0
                  else:
                    ev_equip_power[eqip] = power * round(np.random.uniform(0,0.5),2)
                  ev_hrs[eqip]= ev_equip_power[eqip]/power

                for eqip,power in equip_power.items():
                  if(eqip=='light_1'  or eqip=='light_2'  or eqip=='light_3'  or eqip=='light_4' or eqip=='fan_4'  or eqip=='tv' ):
                    nt_equip_power[eqip] = 0
                  elif eqip=='fridge':
                    nt_equip_power[eqip] = power * round(np.random.uniform(4,5),2)
                  elif eqip=='light_5':
                    nt_equip_power[eqip] = power * round(np.random.uniform(7,8),2)
                  else:
                    nt_equip_power[eqip] = power * round(np.random.uniform(0,8),2)
                  nt_hrs[eqip]= nt_equip_power[eqip]/power
                for eqip,power in base_equipments.items():
                  if(eqip=='ac'):
                    nt_equip_power[eqip] = power * round(np.random.uniform(1,4),2)
                  else:
                    nt_equip_power[eqip] = 0
                  nt_hrs[eqip]= nt_equip_power[eqip]/power


        elif(temp<24):
                if(weekend):
                    for eqip ,power in equip_power.items():
                      if eqip=='fridge':
                        mn_equip_power[eqip] = power * round(np.random.uniform(4,5),2)
                      elif eqip=='light_5':
                        mn_equip_power[eqip] = 0
                      else:
                        mn_equip_power[eqip] = power * round(np.random.uniform(5,6),2)
                      mn_hrs[eqip]= mn_equip_power[eqip]/power
                    for eqip,power in base_equipments.items():
                      if(eqip=='ac'):
                        mn_equip_power[eqip] = 0
                      else:
                        mn_equip_power[eqip] = power * round(np.random.uniform(0,1),2)
                      mn_hrs[eqip]= mn_equip_power[eqip]/power

                    for eqip,power in equip_power.items():
                      if eqip=='fridge':
                        ev_equip_power[eqip] = power * round(np.random.uniform(2,3),2)
                      else:
                        ev_equip_power[eqip] = power * round(np.random.uniform(0,4),2)
                      ev_hrs[eqip]= ev_equip_power[eqip]/power
                    for eqip,power in base_equipments.items():
                      if(eqip=='ac'):
                        ev_equip_power[eqip] = 0
                      else:
                        ev_equip_power[eqip] = power * round(np.random.uniform(0,0.5),2)
                      ev_hrs[eqip]= ev_equip_power[eqip]/power

                    for eqip,power in equip_power.items():
                      if(eqip=='light_1'  or eqip=='light_2'  or eqip=='light_3'  or eqip=='light_4' or eqip=='fan_4'  or eqip=='tv' ):
                        nt_equip_power[eqip] = 0
                      elif eqip=='fridge':
                        nt_equip_power[eqip] = power * round(np.random.uniform(3,4),2)
                      elif eqip=='light_5':
                        nt_equip_power[eqip] = power * round(np.random.uniform(7,8),2)
                      else:
                        nt_equip_power[eqip] = power * round(np.random.uniform(0,7),2)
                      nt_hrs[eqip]= nt_equip_power[eqip]/power
                    for eqip, power in base_equipments.items():
                      nt_equip_power[eqip] = 0
                      nt_hrs[eqip] = 0
                else:

                    for eqip ,power in equip_power.items():
                      if eqip=='fan_3' or eqip=='fan_4':
                        mn_equip_power[eqip] = 0
                      elif eqip=='fridge':
                        mn_equip_power[eqip] = power * round(np.random.uniform(2,3),2)
                      elif eqip=='light_5':
                        mn_equip_power[eqip] = 0
                      else:
                        mn_equip_power[eqip] = power * round(np.random.uniform(0,5),2)
                      mn_hrs[eqip]= mn_equip_power[eqip]/power
                    for eqip,power in base_equipments.items():
                      if(eqip=='ac'):
                        mn_equip_power[eqip] = 0
                      else:
                        mn_equip_power[eqip] = power * round(np.random.uniform(0,1),2)
                      mn_hrs[eqip]= mn_equip_power[eqip]/power

                    for eqip,power in equip_power.items():
                      if eqip=='fridge':
                        ev_equip_power[eqip] = power * round(np.random.uniform(2,3),2)
                      else:
                        ev_equip_power[eqip] = power * round(np.random.uniform(0,4),2)
                      ev_hrs[eqip]= ev_equip_power[eqip]/power
                    for eqip,power in base_equipments.items():
                      if(eqip=='ac'):
                        ev_equip_power[eqip] = 0
                      else:
                        ev_equip_power[eqip] = power * round(np.random.uniform(0,0.5),2)
                      ev_hrs[eqip]= ev_equip_power[eqip]/power

                    for eqip,power in equip_power.items():
                      if(eqip=='light_1'  or eqip=='light_2'  or eqip=='light_3'  or eqip=='light_4' or eqip=='fan_4'  or eqip=='tv' ):
                        nt_equip_power[eqip] = 0
                      elif eqip=='fridge':
                        nt_equip_power[eqip] = power * round(np.random.uniform(3,4),2)
                      elif eqip=='light_5':
                        nt_equip_power[eqip] = power * round(np.random.uniform(7,8),2)
                      else:
                        nt_equip_power[eqip] = power * round(np.random.uniform(0,7),2)
                      nt_hrs[eqip]= nt_equip_power[eqip]/power
                    for eqip, power in base_equipments.items():
                      nt_equip_power[eqip] = 0
                      nt_hrs[eqip] = 0

        mn_power = sum(mn_equip_power.values())
        ev_power= sum(ev_equip_power.values())
        nt_power= sum(nt_equip_power.values())




        mn_unit=round((mn_power/1000),2)
        ev_unit=round((ev_power/1000),2)
        nt_unit=round((nt_power/1000),2)



        mnhrs=sum(mn_hrs.values())
        evhrs=sum(ev_hrs.values())
        nthrs=sum(nt_hrs.values())

        total_unit = mn_unit+ev_unit+nt_unit
        total_hours_used = mnhrs+evhrs+nthrs
        total_power = mn_power + ev_power + nt_power


        def slab_energy_charge(units):

          remaining = units
          bill = 0
          for slab_units, rate in SLABS:
              if remaining <= -1:
                  break
              used = min(remaining, slab_units)
              bill += used * rate
              remaining -= used
          return round(bill, 2)

        def tod_adjustment(mn, ev, nt):


            tod_bill = (
                mn * 0.90 * 9.2 +
                ev * 1.25 * 9.2 +
                nt * 1.00 * 9.2
            )
            return tod_bill

        month_key = (date_obj.year, date_obj.month)

        if month_key != last_month:
            month_unit = 0
            last_month = month_key
            mn_month_unit = 0
            ev_month_unit = 0
            month_unit=total_unit
            nt_month_unit = 0
        else:
            mn_month_unit += mn_unit
            ev_month_unit += ev_unit
            nt_month_unit += nt_unit

            month_unit += total_unit

            monthly_unit[f"{date_obj.month}_{date_obj.year}"] = round(month_unit, 2)

        if month_unit <= 250:
            monthly_bill[f"{date_obj.month}_{date_obj.year}"] = round(slab_energy_charge(month_unit),2)
        elif month_unit <=300:
            monthly_bill[f"{date_obj.month}_{date_obj.year}"] = round(monthly_unit[f"{date_obj.month}_{date_obj.year}"]*6.75,2)
        elif month_unit <=350:
            monthly_bill[f"{date_obj.month}_{date_obj.year}"] = round(monthly_unit[f"{date_obj.month}_{date_obj.year}"]*7.6,2)
        elif month_unit <=400:
            monthly_bill[f"{date_obj.month}_{date_obj.year}"] = round(monthly_unit[f"{date_obj.month}_{date_obj.year}"]*7.95,2)
        elif month_unit <=500:
            monthly_bill[f"{date_obj.month}_{date_obj.year}"] = round(monthly_unit[f"{date_obj.month}_{date_obj.year}"]*8.3,2)
        elif month_unit >500:
            monthly_bill[f"{date_obj.month}_{date_obj.year}"] = round(tod_adjustment(mn_month_unit, ev_month_unit, nt_month_unit),2)

        key=f"{date_obj.month}_{date_obj.year}"


        row = {
              'date': date_str,
              'day_name': day_name,
              'weekend': weekend,
              'temp':temp,
              'total_equipments': num_equip,
              'residents': residents,
              'OP_energy_wh': mn_power,
              'PK_energy_wh': ev_power,
              'NL_energy_wh': nt_power,
              'OP_unit_kwh': mn_unit,
              'PK_unit_kwh': ev_unit,
              'NL_unit_kwh': nt_unit,
              'OP_hrs': mnhrs,
              'PK_hrs': evhrs,
              'NL_hrs': nthrs,
              'total_hours_used': total_hours_used,
              'total_energy_wh': total_power,
              'total_unit_kwh': total_unit,
              'monthly_bill' : monthly_bill.get(key, None),
              'monthly_unit_kwh':  monthly_unit.get(key, None),


          }

        row.update({f'OP_{k}_wh': v for k,v in mn_equip_power.items()})
        row.update({f'PK_{k}_wh': v for k,v in ev_equip_power.items()})
        row.update({f'NL_{k}_wh': v for k,v in nt_equip_power.items()})


        data.append(row)

    df = pd.DataFrame(data)
    df.to_csv('el_data_new.csv', index=False)
    print("CSV file 'el_data_new.csv' has been created.")

    print(df.head())
    print(df.loc[0:31,['date','total_unit_kwh','monthly_bill','monthly_unit_kwh']])
    print(f"Shape: {df.shape} rows")



    from google.colab import files
    #files.download('el_data_new.csv')
```
